```python


# Import the necessary modules
from transformers import pipeline, set_seed

# Define the text generation pipeline
generator = pipeline('text-generation', model='gpt2')

# Set the seed for reproducibility
set_seed(42)

# Generate text with default parameters
print("Default parameters:\n")
print(generator("HuggingFace transformers are", max_length=50, num_return_sequences=1))

# Define a function to experiment with different parameters
def generate_text(prompt, max_length=50, num_return_sequences=1, temperature=1.0, top_k=50):
    return generator(prompt, max_length=max_length, num_return_sequences=num_return_sequences, temperature=temperature, top_k=top_k)

# Example: Experiment with different parameters
prompt = "HuggingFace transformers are"

# Adjust these parameters to see the effect on the generated text
max_length = 100
num_return_sequences = 2
temperature = 0.7
top_k = 30

print(f"\nExperimenting with parameters:\nmax_length={max_length}, num_return_sequences={num_return_sequences}, temperature={temperature}, top_k={top_k}\n")
results = generate_text(prompt, max_length=max_length, num_return_sequences=num_return_sequences, temperature=temperature, top_k=top_k)

# Display the results
for i, result in enumerate(results):
    print(f"Result {i+1}:\n{result['generated_text']}\n")
```
